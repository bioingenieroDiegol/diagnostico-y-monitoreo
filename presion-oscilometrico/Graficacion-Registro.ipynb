```python
import matplotlib.pyplot as plt
import numpy as np
import glob
```

# Lectura del archivo

Primero listamos todos los archivos TXT del directorio actual.

```python
lista = glob.glob('./*.txt')
print(lista)
```

Y ahora mostramos el primero, junto a su tamaño:

```python
archivo = lista[0]
print("Nombre del archivo: ", archivo)
senial = np.loadtxt(archivo)
print("Tamaño del vector: ", np.size(senial))
```

# Graficación

```python
fig, ax = plt.subplots()
ax.plot(senial)

ax.set(xlabel='Elemento', ylabel='Amplitud', title='Señal')
ax.grid()

plt.show()
```
